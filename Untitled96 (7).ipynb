```python
import pandas as pd

df = pd.read_csv("nhanes_clean (1).csv")
df.head()
```

```python
# ===============================
# TARGET: HYPERTENSION (VALID)
# ===============================
df['target'] = (
    (df['sysbp'] >= 140) |
    (df['diabp'] >= 90)
).astype(int)

# ===============================
# FEATURE ENGINEERING
# ===============================
df['age_bp'] = df['age'] * df['diabp']
df['bp_ratio'] = df['sysbp'] / (df['diabp'] + 1)
df['chol_age'] = df['cholesterol'] * df['age']

df['lifestyle_risk'] = (
    (df['smoking'] == 1).astype(int) +
    (df['alcohol'] == 1).astype(int) +
    (df['activity'] == 0).astype(int)
)

df['age_squared'] = df['age'] ** 2
df['chol_log'] = np.log1p(df['cholesterol'])

# ===============================
# REMOVE LEAKAGE FEATURES
# ===============================
X = df.drop(["target","sysbp","diabp"], axis=1)
y = df["target"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)
```

```python
from imblearn.combine import SMOTETomek

sm = SMOTETomek(random_state=42)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [300, 500],
    'learning_rate': [0.01, 0.03],
    'num_leaves': [31, 64],
    'max_depth': [-1, 10],
    'subsample': [0.8, 1.0]
}

grid = GridSearchCV(
    LGBMClassifier(class_weight='balanced'),
    param_grid,
    scoring='roc_auc',
    cv=5,
    n_jobs=-1
)

grid.fit(X_train_sm, y_train_sm)
best_model = grid.best_estimator_
```

```python
from sklearn.calibration import CalibratedClassifierCV

cal_model = CalibratedClassifierCV(best_model, method='isotonic', cv=3)
cal_model.fit(X_train_sm, y_train_sm)

y_prob = cal_model.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import f1_score
import numpy as np

def best_threshold(y_true, y_prob):
    best_t, best_score = 0.5, 0

    for t in np.arange(0.3, 0.9, 0.01):
        y_pred = (y_prob >= t).astype(int)
        score = f1_score(y_true, y_pred)

        if score > best_score:
            best_score = score
            best_t = t

    return best_t

t = best_threshold(y_test, y_prob)
y_pred = (y_prob >= t).astype(int)
```

```python
import numpy as np
from sklearn.metrics import f1_score

def best_threshold(y_true, y_prob):
    best_t, best_score = 0.5, 0

    for t in np.arange(0.3, 0.9, 0.01):
        y_pred = (y_prob >= t).astype(int)
        score = f1_score(y_true, y_pred)

        if score > best_score:
            best_score = score
            best_t = t

    return best_t
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

models = {
    "Logistic": LogisticRegression(max_iter=1000, class_weight='balanced'),
    "RandomForest": RandomForestClassifier(n_estimators=300),
    "SVM": SVC(probability=True, class_weight='balanced'),
    "LightGBM": LGBMClassifier(n_estimators=500, learning_rate=0.03),
}
```

```python
from sklearn.metrics import *

def compute_metrics(y_true, y_pred, y_prob):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

    accuracy = accuracy_score(y_true, y_pred)
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    roc_auc = roc_auc_score(y_true, y_prob)
    precision = precision_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    return accuracy, sensitivity, specificity, roc_auc, precision, f1

results = []

for name, model in models.items():

    # Train
    model.fit(X_train_sm, y_train_sm)

    # Probabilities
    y_prob = model.predict_proba(X_test)[:,1]

    # Optimal threshold
    t = best_threshold(y_test, y_prob)

    # Predictions
    y_pred = (y_prob >= t).astype(int)

    # Metrics
    metrics = compute_metrics(y_test, y_pred, y_prob)

    results.append([name, *metrics])
```

```python
import pandas as pd

evaluation_df = pd.DataFrame(results, columns=[
    "Model",
    "Accuracy",
    "Sensitivity",
    "Specificity",
    "ROC-AUC",
    "Precision",
    "F1-score"
])

# Sort by ROC-AUC (important for paper)
evaluation_df = evaluation_df.sort_values(by="ROC-AUC", ascending=False)

print("\n MODEL PERFORMANCE TABLE:\n")
print(evaluation_df)
```

```python
from sklearn.model_selection import StratifiedKFold
import numpy as np

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize arrays
p1_train = np.zeros(len(X_train_sm))
p2_train = np.zeros(len(X_train_sm))
p3_train = np.zeros(len(X_train_sm))
p4_train = np.zeros(len(X_train_sm))

p1_test = np.zeros(len(X_test))
p2_test = np.zeros(len(X_test))
p3_test = np.zeros(len(X_test))
p4_test = np.zeros(len(X_test))

for train_idx, val_idx in kf.split(X_train_sm, y_train_sm):
    X_tr, X_val = X_train_sm.iloc[train_idx], X_train_sm.iloc[val_idx]
    y_tr, y_val = y_train_sm.iloc[train_idx], y_train_sm.iloc[val_idx]

    # Train base models
    models["Logistic"].fit(X_tr, y_tr)
    models["RandomForest"].fit(X_tr, y_tr)
    models["SVM"].fit(X_tr, y_tr)
    models["LightGBM"].fit(X_tr, y_tr)

    # OOF predictions
    p1_train[val_idx] = models["Logistic"].predict_proba(X_val)[:,1]
    p2_train[val_idx] = models["RandomForest"].predict_proba(X_val)[:,1]
    p3_train[val_idx] = models["SVM"].predict_proba(X_val)[:,1]
    p4_train[val_idx] = models["LightGBM"].predict_proba(X_val)[:,1]

    # Test predictions (average later)
    p1_test += models["Logistic"].predict_proba(X_test)[:,1] / 5
    p2_test += models["RandomForest"].predict_proba(X_test)[:,1] / 5
    p3_test += models["SVM"].predict_proba(X_test)[:,1] / 5
    p4_test += models["LightGBM"].predict_proba(X_test)[:,1] / 5
```

```python
stack_train = np.column_stack((p1_train, p2_train, p3_train, p4_train))
stack_test = np.column_stack((p1_test, p2_test, p3_test, p4_test))
```

```python
#stacking_model
from lightgbm import LGBMClassifier

meta_model = LGBMClassifier(
    n_estimators=300,
    learning_rate=0.03,
    num_leaves=31
)

meta_model.fit(stack_train, y_train_sm)

prob_stack = meta_model.predict_proba(stack_test)[:,1]
```

```python
#weighted_hybrid
from sklearn.linear_model import LogisticRegression

weight_model = LogisticRegression(C=0.5)

weight_model.fit(stack_train, y_train_sm)

prob_weighted = weight_model.predict_proba(stack_test)[:,1]
```

```python
#cost_sensitive
cost_model = LogisticRegression(class_weight={0:3, 1:1})

cost_model.fit(stack_train, y_train_sm)

prob_cost = cost_model.predict_proba(stack_test)[:,1]
```

```python
hybrid_results = []

def evaluate(prob, name):
    t = best_threshold(y_test, prob)
    pred = (prob >= t).astype(int)

    hybrid_results.append([name, *compute_metrics(y_test, pred, prob)])

evaluate(prob_weighted, "Weighted Hybrid")
evaluate(prob_stack, "Stacking")
evaluate(prob_cost, "Cost-Sensitive")

hybrid_df = pd.DataFrame(hybrid_results, columns=[
    "Model","Accuracy","Sensitivity","Specificity","ROC-AUC","Precision","F1"
])

print("\n HYBRID MODELS PERFORMANCE:\n")
print(hybrid_df.sort_values(by="ROC-AUC", ascending=False))
```

```python
# Select best hybrid based on ROC-AUC
best_row = hybrid_df.sort_values(by="ROC-AUC", ascending=False).iloc[0]
print("Best Hybrid:", best_row["Model"])
```

```python
# Assign best probability
if best_row["Model"] == "Weighted Hybrid":
    final_prob = prob_weighted
elif best_row["Model"] == "Stacking":
    final_prob = prob_stack
elif best_row["Model"] == "Cost-Sensitive":
    final_prob = prob_cost
else:
    # Fallback or error handling if model name not matched
    print(f"Warning: Unknown best model '{best_row['Model']}'. Cannot assign final_prob.")
    final_prob = None # Or raise an error

# Final prediction
if final_prob is not None:
    t = best_threshold(y_test, final_prob)
    final_pred = (final_prob >= t).astype(int)

    print("FINAL METRICS:", compute_metrics(y_test, final_pred, final_prob))
else:
    print("Cannot compute final metrics due to unknown model.")
```

```python
from scipy.stats import spearmanr, kendalltau

spearman_corr, _ = spearmanr(shap_rank, lime_rank)
kendall_corr, _ = kendalltau(shap_rank, lime_rank)
```

```python
top_k = 10

shap_top = set(shap_rank[:top_k])
lime_top = set(lime_rank[:top_k])

jaccard = len(shap_top & lime_top) / len(shap_top | lime_top)
```

```python
esi = (spearman_corr + kendall_corr + jaccard) / 3
```

```python
# Step 1: Generate base predictions using threshold
pred = (prob_stack >= best_threshold).astype(int)

# Step 2: Temporary per-sample ESI
esi_sample = np.full(len(prob_stack), esi)

# Step 3: Final predictions with reject option
final_pred = []

for i in range(len(prob_stack)):

    pred_i = pred[i]

    if pred_i == 1:
        if prob_stack[i] < 0.6 or esi_sample[i] < 0.6:
            final_pred.append(-1)
        else:
            final_pred.append(1)

    else:  # pred_i == 0
        if prob_stack[i] > 0.4 or esi_sample[i] < 0.6:
            final_pred.append(-1)
        else:
            final_pred.append(0)

# Step 4: Convert to array
final_pred = np.array(final_pred)
```
